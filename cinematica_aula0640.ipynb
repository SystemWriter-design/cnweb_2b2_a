# Fórmula da velocidade
fonte: [https://www.todamateria.com.br/formulas-de-cinematica/]

```
def calculaVelMedia(deltaS, deltaT):
  velMedia = (deltaS/deltaT)
  return velMedia
```

## Teste Serviço Nuvem: Velocidade Média

```
print("Vm=", calculaVelMedia(30, 15))
```

# Fórmulas do Movimento Uniforme (MU)

```
def calculaMU(deltaS, deltaT):
  MU = (deltaS/deltaT)
  return MU
```

# Função horária do deslocamento

```
def calculaDesloc(inicialS, deltaT, v):
  d = inicialS + (v*deltaT)
  return d
```

## Testes de Função : Horária de Deslocamento

```
print("Deslocamento= ",calculaDesloc(0, 15, 30))
```

## Função horária da velocidade

```
def calculaFuncaoVel(v0, a, t):
  v = v0 + (a*t)
  return v
```

## Teste horária da velocidade

```
print("Função horária: ", calculaFuncaoVel())
```

## Função horária da posição

```
import math
def calculaFuncaoPos(v0, a, t):
  s = v0*t + round(math.sqrt(a*(t**2)/2))
  return s
```

# Teste horária da posição

```
Print("Função posição: ", calculaFuncaoPos())
```

# Equação de Torricelli

```
import math

def calTorricelli(vZero, acel, des):
  t=2*acel*des
  v=vZero+t
  res = round(math.sqrt(v), 2)
  return res
```

# Teste de Torricelli

```
print("Calculo de Torriceli: " ,calTorricelli(0, 1, 8))
```
